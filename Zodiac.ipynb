```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
zodiac_traits = {
    "Aries": "Courageous, determined, confident",
    "Taurus": "Reliable, patient, practical",
    "Gemini": "Gentle, affectionate, curious",
    "Cancer": "Tenacious, highly imaginative, loyal",
    "Leo": "Creative, passionate, generous",
    "Virgo": "Loyal, analytical, kind",
    "Libra": "Cooperative, diplomatic, gracious",
    "Scorpio": "Resourceful, brave, passionate",
    "Sagittarius": "Generous, idealistic, great sense of humor",
    "Capricorn": "Responsible, disciplined, self-control",
    "Aquarius": "Progressive, original, independent",
    "Pisces": "Compassionate, artistic, intuitive"
}
```

```python
questions = [
    "What is your birth month (1-12)?",
    "What is your birth day (1-31)?",
    "Do you consider yourself more introverted or extroverted?",
    "Do you prefer stability or spontaneity?",
    "Would you say you are more logical or emotional?",
    "Do you find yourself more drawn to artistic or scientific pursuits?"
]
```

```python
def predict_zodiac(answers):
    month = int(answers[0])
    day = int(answers[1])

    if (month == 1 and day <= 19) or (month == 12 and day >= 22):
        return "Capricorn"
    elif (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return "Aquarius"
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return "Pisces"
    elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return "Aries"
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return "Taurus"
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return "Gemini"
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return "Cancer"
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return "Leo"
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return "Virgo"
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return "Libra"
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return "Scorpio"
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return "Sagittarius"
    return "Unknown"  # Just in case something goes wrong

def get_personality_traits(zodiac_sign):
    return zodiac_traits[zodiac_sign]
```

```python
def collect_user_input():
    answers = []
    for question in questions:
        answer = input(question + " ")
        answers.append(answer)
    return answers
```

```python
def chatbot():
    print("Welcome to the Zodiac Sign Predictor!")
    answers = collect_user_input()
    zodiac_sign = predict_zodiac(answers)
    traits = get_personality_traits(zodiac_sign)
    print(f"Your zodiac sign is {zodiac_sign}.")
    print(f"Major personality traits: {traits}")

chatbot()
```
